```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
#now applying the OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
#applying gradient decent assuming slope is constant m =78.35
#and lets assume the starting value for intercept b= 0
y_pred = ((78.35 * X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label = 'OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# main process
m = 78.35
b= 0
loss_slope = -2 * np.sum(y - m*X.ravel() -b )
```

```python
loss_slope
```

```python
# now taking the learning rate
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept
b= b -step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
#using a loop function for optimizing model through the gradient decent
b = 100
m = 78.35
lr = 0.01

epochs = 100
for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b) # ravel function lae flatten garxa array lai 1D ma
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
